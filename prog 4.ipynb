```python
# Input tuple
t = (1, 2, 2, 3, 1, 4)

#for frequency of the element
frequency = {}

for item in t:
    if item in frequency:
        frequency[item] += 1
    else:
        frequency[item] = 1

#print the result
print("Frequency:", frequency)
```
