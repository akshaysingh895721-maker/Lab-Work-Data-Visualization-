```python
# Input the list
numbers = [1, 2, 3, 2, 4, 5, 1]

duplicates = []
data = set()

for num in numbers:
    if num in data:
        if num not in duplicates:
            duplicates.append(num)
    else:
        data.add(num)

print("Duplicate elements:", duplicates)
```
