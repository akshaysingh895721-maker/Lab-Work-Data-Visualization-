```python
#find the maximum and minimum element in tuple

# Input tuple
t = (5, 1, 8, 3)

# Initialize max and min with the first element
max_val = t[0]
min_val = t[0]

# Traverse the tuple
for num in t:
    if num > max_val:
        max_val = num
    if num < min_val:
        min_val = num

print("Max =", max_val, ", Min =", min_val)
```
